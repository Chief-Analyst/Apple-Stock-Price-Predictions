```python
import pandas as pd
import yfinance as yf
```

```python
##Dowload Dataset

ticker = yf.Ticker("AAPL")

data = ticker.history(period="10y", interval="1d")

print(data.head())
```

```python
print(data.shape)
print(data.index[:5])
print(data.columns)
```

```python
import numpy as np
import tensorflow as tf
```

```python
data.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.callbacks import EarlyStopping
```

```python
##Map out features and targets

features = data[["Open", "High", "Low", "Volume"]].values

targets = data["Close"].values.reshape(-1,1)
```

```python
##Scale features and targets

feature_scaler = MinMaxScaler()
features_scaled = feature_scaler.fit_transform(features)

target_scaler = MinMaxScaler()
targets_scaled = target_scaler.fit_transform(targets) 
```

```python
##Create Sequence for LSTM model

SEQ_LEN = 60

def create_sequences(features, targets, seq_length = SEQ_LEN):
    X, y = [], []
    for i in range(seq_length, len(features)):
        X.append(features[i-seq_length: i])
        y.append(targets[i,0])

    return np.array(X), np.array(y)

X, y = create_sequences(features_scaled, targets_scaled)
                 
    
  
```

```python
##Split data into Train, Validation and Test 

train_size = int(len(X) * 0.7)
val_size = int(len(X) * 0.15)

X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size: train_size + val_size], y[train_size: train_size + val_size]
X_test, y_test = X[train_size + val_size:], y[train_size + val_size:]


```

```python
##Build Model

model = Sequential()
model.add(LSTM(50, return_sequences = True, input_shape = (SEQ_LEN, X.shape[2])))
model.add(LSTM(50))
model.add(Dense(1))

model.compile(optimizer = "adam", loss= "mean_squared_error")
early_stop = EarlyStopping(monitor = "val_loss", patience = 10, restore_best_weights = True)
```

```python
##Train Model

history = model.fit(
    X_train, y_train,
    epochs = 50,
    batch_size = 32,
    validation_data = (X_val, y_val), 
    callbacks = [early_stop])
    
```

```python
##Predict on test set

import matplotlib.pyplot as plt

y_pred_scaled = model.predict(X_test)

##Inverse scale target Close
y_pred = target_scaler.inverse_transform(y_pred_scaled)
y_true = target_scaler.inverse_transform(y_test.reshape(-1,1))

# Plot
plt.figure(figsize=(12,6))
plt.plot(y_true, color='blue', label='Actual Close Price')
plt.plot(y_pred, color='red', label='Predicted Close Price')
plt.title('AAPL Close Price Prediction with OHLCV')
plt.xlabel('Time')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
##evaluate Model Performance

from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

##RMSE
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f"RMSE: {rmse:.2f} USD")

##MAPE
mape = mean_absolute_percentage_error(y_true, y_pred) * 100
print(f"MAPE: {mape:.2f}%")
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_true, color='blue', label='Actual Close')
plt.plot(y_pred, color='red', label='Predicted Close')
plt.fill_between(range(len(y_true)),
                 (y_true - rmse).flatten(),
                 (y_true + rmse).flatten(),
                 color='red', alpha=0.1, label='RMSE Band')
plt.title('AAPL Close Price Prediction with Error Band')
plt.xlabel('Time')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

USING THE MODEL TO PREDICT PRICE FOR THE NEXT FIVE DAYS


```python
from tensorflow.keras.models import load_model

model = load_model("AAPL_LSTM_model.keras")
```

```python
import numpy as np

last_sequence = features_scaled[-SEQ_LEN:]  
last_sequence = np.expand_dims(last_sequence, axis=0)
```

```python
def predict_future(model, last_sequence, n_days, target_scaler):
    predictions = []
    current_seq = last_sequence.copy()

    for _ in range(n_days):
        next_price_scaled = model.predict(current_seq, verbose=0)
        next_price = target_scaler.inverse_transform(next_price_scaled)[0,0]
        predictions.append(next_price)

        ##Create next input step
        next_step = current_seq[0, -1].copy()
        next_step[3] = next_price_scaled[0,0]  # index 3 = Close

        ##Roll sequence forward
        current_seq = np.roll(current_seq, -1, axis=1)
        current_seq[0, -1] = next_step

    return predictions
```

```python
future_days = 5
future_prices = predict_future(
    model,
    last_sequence,
    future_days,
    target_scaler
)

print("Next 5 predicted close prices:")
for i, price in enumerate(future_prices, 1):
    print(f"Day {i}: ${price:.2f}")
```

```python
##To add dates for better readability

last_date = data.index[-1]
future_dates = pd.bdate_range(
    start=last_date + pd.Timedelta(days=1),
    periods=5
)

for d, p in zip(future_dates, future_prices):
    print(d.date(), f"${p:.2f}")
```
